## Loading Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Import data

```python
df = pd.read_csv("top50.csv", engine = 'python')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
df.head()
```

```python
df['Artist.Name'].value_counts()
```

```python
df['Genre'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy(deep=True)
enc = LabelEncoder()
df_encoded.loc[:,['Artist.Name','Genre']] = df_encoded.loc[:,['Artist.Name','Genre']].apply(enc.fit_transform)
df_encoded.head()
```

```python
df_to_cluster = df_encoded.loc[:,'Artist.Name':] 
df_to_cluster.head()
```

```python
from sklearn.preprocessing import StandardScaler
stdscaler = StandardScaler()

scaled_data = pd.DataFrame(stdscaler.fit_transform(df_to_cluster), index=df_to_cluster.index, columns=df_to_cluster.columns)
scaled_data.head()
```

```python
sns.heatmap(scaled_data.corr(),vmin=-1,vmax=1,center=0,cmap='coolwarm')
```

```python
## considering popularity and genre
data = scaled_data[['Genre','Popularity']]
data.head()
```

```python
# find the appropriate cluster number
plt.figure(figsize=(10, 8))
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 20):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 20), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
## Clustering data based on Genre and Popularity
# Fitting K-Means to the dataset
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 6, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(scaled_data)

from sklearn.metrics import silhouette_score
df['Cluster_label'] = pd.DataFrame(y_kmeans)
```

```python
df.head()
```

```python
val = df.groupby('Cluster_label').agg('mean')
val.sort_values('Popularity',ascending=False) ##Sorting By Popularity
```

```python
genre = df[df['Cluster_label']==4].Genre
genre.head()
```

### This shows, based on the Genre data is divided into 6 clusters, and based on popularity, mean beats.per.min is 136.666667.
### Also, Genres like these -> electropop,dance pop,dfw rap,trap music,pop tend to gain more popularity.(Considering Cluster 4 and 5)

```python
ax=sns.swarmplot(x="Cluster_label",y="Genre",data=df)
```

```python
ax=sns.swarmplot(x="Cluster_label",y="Popularity",data=df)
```

```python
#sns.pairplot(df,hue='Cluster_label')
```

```python
silhouette_avg = silhouette_score(scaled_data, df["Cluster_label"])
silhouette_avg
```

```python
from sklearn.metrics import davies_bouldin_score
davies_bouldin_score(scaled_data, df["Cluster_label"])
```

## Using PCA on scaled data and then performing K-Means

```python
scaled_data.head()
```

```python
from sklearn.decomposition import PCA

pca_model = PCA(n_components = 2 )
pca_data = pd.DataFrame(pca_model.fit_transform(scaled_data), index=scaled_data.index, columns=['PC-1','PC-2'])
```

```python
pca_data.head()
```

```python
pca_data.shape
```

```python
# find the appropriate cluster number
plt.figure(figsize=(10, 8))
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 20):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(pca_data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 20), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans_pca = KMeans(n_clusters = 6, init = 'k-means++', random_state = 42)
y_kmeans_pca = kmeans_pca.fit_predict(pca_data)

from sklearn.metrics import silhouette_score
df['Cluster_label_pca'] = pd.DataFrame(y_kmeans_pca)
```

```python
df.head()
```

```python
silhouette_avg = silhouette_score(scaled_data, df["Cluster_label_pca"])
silhouette_avg
```

```python
davies_bouldin_score(scaled_data, df["Cluster_label_pca"])
```

```python
val_pca = df.groupby('Cluster_label_pca').agg('mean')
val_pca.sort_values('Popularity',ascending=False) ##Sorting By Popularity
```

```python
## hirerachial
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(pca_data, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('PCA data')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
##considering clusters based on dendogram data
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(pca_data)
```

```python
df['Cluster_label_hc'] = pd.DataFrame(y_hc)
```

```python
df.head()
```

```python
val_hc = df.groupby('Cluster_label_hc').agg('mean')
val_hc.sort_values('Popularity',ascending=False) ##Sorting By Popularity
```

```python
silhouette_avg_hc = silhouette_score(scaled_data, df["Cluster_label_hc"])
silhouette_avg_hc
```

```python
davies_bouldin_score(scaled_data, df["Cluster_label_hc"])
```

```python
genre = df[df['Cluster_label_hc']==1].Genre
genre.head()
```
